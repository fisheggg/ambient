```python
# Reading time, light intensity and gain from the csv


# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(mobile_data, header=1,  delimiter=';',decimal=',', low_memory=False)

# Specify the column names you want to display
selected_columns = ['time', 'I', 'Gain']

# Select and display the specified columns
df_selected = df[selected_columns]
print(df_selected)
```

```python
df_selected
```

```python
df_selected.dtypes
```

```python
# Replacing n-dash with hyphen
df_selected=df_selected.replace(to_replace='−', value='-',regex=True)
# Replacing comma with dot
df_selected=df_selected.replace(to_replace=',', value='.',regex=True)
# Replacing NAN with zeros
df_selected=df_selected.replace(to_replace='∞', value='0',regex=True)

df_selected
```

```python
# Now that the data should have been formatted correctly, we can change to float64
df_selected=df_selected.astype(float)
df_selected.dtypes
```

```python
#Calculating the sampling rate

total_duration = df_selected['time'].max() - df_selected['time'].min()
num_of_data_points = len(df_selected)

sr = num_of_data_points / total_duration

print(total_duration)
sr
```

```python
#The sampling rate is higher than necessary.
#Therefore, we remove rows with no change in data as they are redundant.
#We use Gain here because it fluctuates more than light intensity, so we end up losing the least data.

df_selected = df_selected.drop_duplicates(subset=['Gain'], keep='first')
df_selected
```

```python
#Calculating the new sampling rate

total_duration = df_selected['time'].max() - df_selected['time'].min()
num_of_data_points = len(df_selected)

sr = num_of_data_points / total_duration

print(total_duration)
sr
```
